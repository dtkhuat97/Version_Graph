```python
from subprocess import call
import csv
import os

path_og = "/home/dtkhuat/preprocessed_files/final/"
path_sizes = "/home/dtkhuat/results_version/size_"
path_incidence =  "/home/dtkhuat/results_version/itr/real_graphs_compressed/"
path_index = "/home/dtkhuat/results_version/version_itr/real_graphs_compressed/"


#date = "26_04_24"
date = "real_12_06_2024"

file_num = "2"

file_name = "allData_p" + file_num
compressed_name = file_name + "_compressed"
    
    
with open('/home/dtkhuat/results_version/'+date+'.csv', 'a') as csv_file:
    csv_file.write(file_name+", ")
    csv_file.close()
print("START ITR COMPRESSION")
call(["/home/dtkhuat/IncidenceTypeRePair/build/./cgraph-cli", "--overwrite", "--verbose","--format", "hyperedge", "--max-rank", "64", path_og + file_name,  path_incidence+compressed_name])

with open('/home/dtkhuat/results_version/'+date+'.csv', 'a') as csv_file:
    csv_file.write(", ")
    csv_file.close()

print("START VERSION ITR COMPRESSION")
call(["./cgraph-cli", "--overwrite", "--verbose","--format", "hyperedge", "--max-rank", "64" ,path_og+file_name, path_index + compressed_name])

with open('/home/dtkhuat/results_version/'+date+'.csv', 'a') as csv_file:
    csv_file.write("\n")
    csv_file.close()
```

```python
from subprocess import call
import csv
import os

path_og = "/home/dtkhuat/eval_datasets/semi_real_datasets/"
path_size = "/home/dtkhuat/results_version/size_"
path_incidence =  "/home/dtkhuat/results_version/itr/semi_real_graphs_compressed/"
path_index = "/home/dtkhuat/results_version/version_itr/semi_real_graphs_compressed/"


#date = "26_04_24"
#date = "semi_real_13_06_2024"
date = "semi_real_18_06_2024"
#file_num = "2"


    
for file_num in range (2,5):
    file_name = "lubm-10" + str(file_num)
    #file_name = "taxonomy" + str(file_num)
    compressed_name = file_name + "_compressed"
    with open('/home/dtkhuat/results_version/'+date+'.csv', 'a') as csv_file:
        csv_file.write(file_name+", ")
        csv_file.close()
    print("START ITR COMPRESSION")
    call(["/home/dtkhuat/IncidenceTypeRePair/build/./cgraph-cli", "--overwrite", "--verbose","--format", "hyperedge", "--max-rank", "64", path_og + file_name,  path_incidence+compressed_name])

    with open('/home/dtkhuat/results_version/'+date+'.csv', 'a') as csv_file:
        csv_file.write(", ")
        csv_file.close()

    print("START VERSION ITR COMPRESSION")
    call(["./cgraph-cli", "--overwrite", "--verbose","--format", "hyperedge", "--max-rank", "64" ,path_og+file_name, path_index + compressed_name])


    with open(path_size+date+'.csv', 'a') as csv_file:
        fields = [file_name, os.stat(path_og + file_name).st_size, os.stat(path_incidence+compressed_name).st_size, os.stat(path_index+compressed_name).st_size]
        csv_writer = csv.writer(csv_file, delimiter=',')
        csv_writer.writerow(fields)
    with open('/home/dtkhuat/results_version/'+date+'.csv', 'a') as csv_file:
        csv_file.write("\n")
        csv_file.close()
```
