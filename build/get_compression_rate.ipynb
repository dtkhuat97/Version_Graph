```python
from subprocess import call
import csv
import os
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

#PLOT C RATE FOR MAXIMUM LABEL SIZE


```python
path = "/home/dtkhuat/results/random_graphs/var_edge_len/"
str_len = 4
increment = 4
c_size = []
original = []
ir_c = []
ie_c = []
while str_len < 65:
    temp = []
    temp2 = []
    temp3 = []
    with open(path+str(str_len)+"/"+'23_04_24.csv', 'r') as csv_file:
        csvreader = csv.reader(csv_file)
        for row in csvreader:
            temp.append(float(row[1]))
            temp2.append(float(row[2]))
            temp3.append(float(row[3]))
    original.append(temp)
    ir_c.append(temp2)
    ie_c.append(temp3)
    str_len = str_len + increment

fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
plt.xlabel("Maximum label size of node and edges")
plt.ylabel("Compression ratio")
#bp = ax.stem(timings,showfliers=False,showmeans=True)
x = []
y = []

for i in range(0,16):
    y.append(statistics.mean(ir_c[i])/statistics.mean(original[i]))
    x.append((i+1)*4)
bp = ax.bar(x, y, linewidth=2, capsize=6)
plt.savefig("/home/dtkhuat/plots/c_rate/ir_var_edge_len2.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.close()

fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
plt.xlabel("Maximum label size of node and edges")
plt.ylabel("Compression ratio")
x = []
y = []

for i in range(0,16):
    y.append(statistics.mean(ie_c[i])/statistics.mean(original[i]))
    x.append((i+1)*4)
bp = ax.bar(x, y, linewidth=2, capsize=6)
plt.savefig("/home/dtkhuat/plots/c_rate/ie_var_edge_len2.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.close()
```

#PLOT C RATE REAL GRAPHS

```python

path = "/home/dtkhuat/results/sizes/"

original = []
ir_c = []
ie_c = []
x = []
y = []
with open(path+'21_04_24.csv', 'r') as csv_file:
    csvreader = csv.reader(csv_file)
    next(csvreader, None) 
    for row in csvreader:
        x.append(row[0])
        original.append(float(row[1]))
        ir_c.append(float(row[2]))
        ie_c.append(float(row[3]))

x=x[:-1]
print(np.asarray(original).shape)
fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
plt.xlabel("File Name")
plt.ylabel("Compression ratio")

y = []
for i in range(0,12):
    y.append(ir_c[i]/original[i])
bp = ax.bar(x, y)
plt.xticks(rotation=90)
plt.savefig("/home/dtkhuat/plots/real_graphs/c_rate/ir_c_rate.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.close()

fig = plt.figure(figsize =(5, 5))
ax = fig.add_axes([0, 0, 1, 1])
plt.xlabel("File Name")
plt.ylabel("Compression ratio")

y = []
for i in range(0,12):
    y.append(ie_c[i]/original[i])
bp = ax.bar(x, y)
plt.xticks(rotation=90)
plt.savefig("/home/dtkhuat/plots/real_graphs/c_rate/ie_c_rate.pdf", format="pdf", bbox_inches="tight")
plt.show()
plt.close()


```
